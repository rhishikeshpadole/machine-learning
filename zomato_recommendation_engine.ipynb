```python
import pandas as pd
import numpy as np
import surprise
```

```python
ratings = pd.read_csv("https://raw.githubusercontent.com/rhishikeshpadole/web-scraped-datasets/master/zomato_dataset.csv", sep = ',')
ratings
```

```python

from surprise import Reader, Dataset, KNNWithMeans
reader = Reader(rating_scale=(1,5))

data = Dataset.load_from_df(ratings[['UserName','Restro_name','Rating']], 
                           reader)
```

```python
sim_options = {'user_based':True, 'name':'cosine'}
model = KNNWithMeans(sim_options=sim_options)
model.fit(data.build_full_trainset())
```

```python
predicted_ratings = pd.DataFrame(columns = ['UserName','Restro_name','Rating'])
for j in ratings['UserName'].unique():
    for i in ratings['Restro_name'].unique():
        user_mid = ratings[(ratings['UserName'] == j) & 
                     (ratings['Restro_name'] == i)]
        is_rated = True if len(user_mid) > 0 else False
  
    if not is_rated:
      
      curr = {'UserName': j , 
           'Restro_name' : i ,
           'Rating' : model.predict(j, i)[3]}
        predicted_ratings = predicted_ratings.append(curr,ignore_index=True)
```

```python
recommend_df = predicted_ratings.sort_values(by = 'Rating', ascending = False).groupby('UserName').head(3)
```
